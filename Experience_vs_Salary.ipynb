```python
import matplotlib.pyplot as plt

plt.scatter(salary_data["YearsExperience"],salary_data["Salary"], color='green')
plt.title(" Exp vs salary")
plt.xlabel("years of experience")
plt.ylabel("salary (INR)")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = salary_data[["YearsExperience"]]
y = salary_data[["Salary"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
print("if a person is 15 of exp......")
print("predicted salary: ", model.predict([[1]])[0])
```

```python
from sklearn.linear_model import LinearRegression
import joblib

joblib.dump(model, "salary_model.pkl")
print("Model saved as salary_model.pkl")
```
